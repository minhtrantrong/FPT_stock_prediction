```python
from bs4 import BeautifulSoup
import pandas as pd
import lxml
import requests
```

```python
url='https://s.cafef.vn/Lich-su-giao-dich-FPT-1.chn'
#Create a handle, page, to handle the contents of the website
page = requests.get(url).text
# Fetch the page text to BeautifulSoup
soup = BeautifulSoup(page, 'lxml')
```

```python
# check table
all_tr = soup.find_all('tr')
```

```python
for tr in all_tr:
    all_td = tr.find_all('td')
    for td in all_td:
        print(td.text)
```

```python
i = 1
for td in all_tr[3].find_all('td'):
    data=td.text
    print(i, ": ", data)
    i+=1
```

```python
# list all col name
col_list = ["Ngay", "Gia d/c", "Gia dong cua", "Thay doi", "up-down",
            "GD-KL", "GD-GT", "TT-KL", "TT-GT", "Gia mo cua", 
            "Gia cao nhat", "Gia thap nhat", "C13", "C14", "C15" ]
#Create empty list
col=[]
#For each row, store each first element (header) and an empty list
for t in col_list:
    col.append((t,[]))

print(col, len(col))
```

```python
#Since 3 first row is the header, data is stored on the 4th row onwards
for j in range(3,len(all_tr)):
    #row is our j'th row
    row=all_tr[j]
    all_td = row.find_all('td')
    
    #i is the index of our column
    i=0
    
    #Iterate through each element of the row
    for td in all_td:
        data=td.text 
        if i>14: 
            print(all_td, len(all_td))
            continue
        #Check if row is empty
        if i>0:
            #print(i)
        #Convert any numerical value to integers
            try:
                data=int(data)
            except:
                pass
        #Append the data to the empty list of the i'th column
        col[i][1].append(data)
        #Increment i for the next column
        i+=1
```
